```python
import sqlite3
import pandas as panda
from pprint import pp

# Connexion à la base de données (si elle n'existe pas, elle sera créée)
conn = sqlite3.connect('data/databases/database.db')

# Création d'un curseur pour exécuter des commandes SQL
cur = conn.cursor()

# Nom de la table
table_name = "episode"

# Définition du schéma de la table
cur.execute('''
    CREATE TABLE IF NOT EXISTS episode (
        id INTEGER PRIMARY KEY,
        name TEXT,
        episode_number INTEGER,
        episode_season INTEGER,
        date DATE,
        country TEXT,
        channel TEXT
    )
''')

# Validation des modifications
try:
    conn.commit()
except:
    print("Erreur dans la création de la bdd.")
    conn.rollback()

# Ajout d'une donnée dans la table
query = panda.DataFrame([{
        'name': 'Test',
        'episode_number': 1,
        'episode_season': 1,
        'date': '03-11-2023',
        'country': 'France',
        'channel': 'BFMTV'
    }])

query.to_sql(
    name=table_name, 
    con=conn, 
    if_exists='append', 
    index=False, 
    chunksize=None
    )

# Lecture de la table (df -> dataframe)
df = panda.read_sql(f'SELECT * FROM {table_name}', con=conn)

# Obtenir une liste de listes : 
data = [
    df[col].tolist()
    for col in df.columns
]

pp(data)
```
